Se importan las librerias a utilizar


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import altair as alt
```

```python
df = pd.read_csv('/content/Advertising.csv')
df.columns = ['Caso','TV', 'Radio', 'Periodico', 'ventas']
df.head(10)
```

```python
df_melted = df.melt(id_vars=['Caso', 'ventas'], value_vars=['TV', 'Radio', 'Periodico'], var_name='Variable', value_name='Valor')
df_melted.head(10)
chart = alt.Chart(df_melted).mark_circle().encode(
    x='Valor',
    y='ventas',
    color='Variable'
).properties(
    title='TV, Radio, y Periodico vs Sales'
)
chart
```
